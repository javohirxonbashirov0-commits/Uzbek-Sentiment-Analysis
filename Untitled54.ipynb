```python
import torch
import torch.nn as nn
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    Trainer,
    TrainingArguments,
    pipeline
)
from datasets import Dataset
device ="cuda" if torch.cuda.is_available() else "cpu"
```

```python
data = {
    "text":[
        "Bu mahsulot juda ajoyib, menga yoqdi!",
        "Sifati juda yomon, tavsiya qilmayman.",
        "Yetkazib berish tez bo'ldi, rahmat.",
        "Narxi juda qimmat, arzimaydi.",
        "Darslar juda tushunarli ekan."
    ],
    "label":[1,0,1,0,1]
}
```

```python
dataset = Dataset.from_dict(data)
dataset =dataset.train_test_split(test_size=0.2)
```

```python
model_nm = "xlm-roberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_nm)

def tokenize_func(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)

tokenized_datasets = dataset.map(tokenize_func, batched=True)
```

```python
model =AutoModelForSequenceClassification.from_pretrained(model_nm,num_labels=2).to(device)
training_args = TrainingArguments(
    output_dir="./xlm_roberta_uzbek_sentiment",
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    num_train_epochs=10,
    weight_decay=0.01,
    logging_dir='./logs',
    fp16=True if device == "cuda" else False,
    logging_steps=5,
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset =tokenized_datasets["train"],
    eval_dataset = tokenized_datasets["test"],
)
trainer.train()
```

```python
trainer.save_model("./xlm_roberta_uzbek_sentiment")
tokenizer.save_pretrained("./xlm_roberta_uzbek_sentiment")
test_sentence = "Bugungi ob-havo juda ajoyib va yoqimli"
post_pipe =pipeline("sentiment-analysis",model="./xlm_roberta_uzbek_sentiment",tokenizer=model_nm)
print(post_pipe(test_sentence))
```
